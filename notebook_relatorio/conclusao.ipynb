Foi feita uma verificação para saber a correlação entre o tamanho da sinopse do filme e a nota do filme, foram considerados apenas filmes com nota maior ou igual a 8.3 (nota média dos 250 filmes). 
O resultado foi que a correlação entre o tamanho da sinopse e a nota do filme é 0.05, sendo considerada uma correlação fraca, indicando que o tamanho da sinopse não influencia a nota do filme.
